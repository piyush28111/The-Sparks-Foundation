## Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('globalterrorismdb_0718dist.csv',encoding='ISO-8859-1')
data.head()
```

## Renaming the columns

```python
data= data.rename(columns={'iyear':'year','imonth':'month','iday':'day','country_txt':'Country','region_txt':'Region','provstate':'state','attacktype1_txt':'Attack_type','targtype1_txt':'Target_type','targsubtype1_txt':'Target_subtype','weaptype1_txt':'weapon_type','weapsubtype1_txt':'weapon_subtype','weapdetail':'weapon_detail','propextent_txt':'property_extent','nkill':'killed','nwound':'wounded'})
data.head()
```

```python
pd.set_option('Display.max_columns',None)
```

```python
data=data[['year','month','day','extended','Country','Region','state','city','latitude','longitude','summary','Attack_type','Target_type','Target_subtype','motive','weapon_type','weapon_subtype','weapon_detail','killed','wounded','property_extent']]
data.head()
```

## Filling Null values as 0

```python
data=data.fillna(0)
data.head()
```

# Analysis of data


```python
plt.figure(figsize=(6,4),dpi=120)
data.Country.value_counts().sort_values(ascending=False).head(20).plot(kind='bar',color=(1,0.4,0.4))
plt.title('Top 20 affected countries by terrorism')
plt.xlabel('Countries')
plt.ylabel('Number of times affected')
plt.show()
```

### <i>Iraq is most affected country by terrorism</i>

```python
plt.figure(figsize=(6,4),dpi=120)
data.state[data.state!='Unknown'].value_counts()[:20].plot(kind='bar',color=(0.2,0.7,0.8))
plt.title('Top 20 affected states by terrorism')
plt.xlabel('States')
plt.ylabel('Number of times affected')
```

### <i>Baghdad is most affected state by terrorism </i>

```python
plt.figure(figsize=(8,5),dpi=100)
sns.countplot('Region',data=data)
plt.xticks(rotation=90)
plt.title('Affected regions')
```

### <i>Most terrorist attacks were made in Middle East & North Africa</i>

```python
r=pd.DataFrame(data.Region.value_counts())
fig,ax = plt.subplots(dpi=100,figsize=(9,7))
ax.pie(r.Region,labels=r.index,autopct='%1.1f%%')
```

### <i> Percentage of total attacks in different regions</i>

```python
plt.figure(figsize=(8,5),dpi=120)
sns.countplot('year',data=data)
plt.xticks(rotation=90)
plt.title('Number of terrorist attacks per year')
```

### <i>Most number of terrorist attacks held in 2014</i>

```python
plt.figure(figsize=(8,4),dpi=100)
sns.countplot('month',data=data)
plt.xticks(rotation=0)
plt.title('Number of terrorist attacks per month')
```

### <i>No specific month was targeted by terrorist but most of the attacks were in May</i>

```python
plt.figure(figsize=(9,5),dpi=100)
sns.countplot('Attack_type',data=data,palette='PuBu')
plt.xticks(rotation=90)
plt.title('Attack type')
```

### <i> Most terrorist attacks were made by Bombing and Explosions</i>

```python
plt.figure(figsize=(9,6),dpi=100)
sns.countplot('Target_type',data=data,palette='YlGnBu')
plt.xticks(rotation=90)
plt.title('Target type')
```

### <i>Most of the attacks were made on to Private Citizens & property</i>

```python
plt.figure(figsize=(7,4),dpi=120)
data.Target_subtype[(data.Target_subtype!='Unknown') & (data.Target_subtype!=0)].value_counts()[:20].plot(kind='bar',color=(0.7,0.5,0.5))
plt.title('Top 20 Target_subtype')
plt.xlabel('Taget_subtype')
plt.ylabel('Number of times affected')
```

### <i>Unnamed civilians and police security forces were made major targets</i> 

```python
plt.figure(figsize=(8,5),dpi=100)
sns.countplot(data[data['weapon_type']!='Unknown']['weapon_type'],data=data)
plt.xticks(rotation=90)
```

### <i> Attacks were majorly made by Explosives weapons </i>

```python
plt.figure(figsize=(9,6),dpi=100)
sns.countplot(data[data['weapon_subtype']!=0]['weapon_subtype'],data=data)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(7,4),dpi=120)
data.weapon_detail[data.weapon_detail!=0].value_counts()[:20].plot(kind='bar',color=(0.7,0.5,0.5))
plt.title('Top 20 weapon details')
plt.xlabel('weapon_detail')
plt.ylabel('Number of times used')
```

### <i> Attacks were made by explosives followed by automatic firearm </i>

## Casualities are total number of people got affected by attacks , it included both killed and        wounded

```python
data['casualities']=data['killed']+data['wounded']
```

### <i> Total data of casualities </i>

```python
pd.DataFrame({'Killed':data['killed'].sum(),'wounded':data['wounded'].sum(),'casualities':data['casualities'].sum()},index=['Total'])
```

```python
plt.figure(figsize=(6,3),dpi=120)
sns.countplot(data[(data['property_extent']!=0) & (data['property_extent']!='Unknown')]['property_extent'])
plt.xticks(rotation=30)
```

### <i> Minor(likely < 1 million dollars) property was affected most by these attacks

# Grouping the data 

```python
dt = data[(data['state']!=0) & (data['state']!='Unknown')]
grp = dt.groupby(['year','Country','state','Target_type']).casualities.sum()
```

### <i> It gives a detailed analysis of at what particular year in which state who was targeted with this much amount of casualities
