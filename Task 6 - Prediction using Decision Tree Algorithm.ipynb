## Importing libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
import seaborn as sns
```

## Reading the file 

```python
data = pd.read_csv('Iris.csv')
data
```

## Spiltting the data into 2 variables

```python
y = data['Species']
x=data.drop(['Id','Species'],axis=1)
```

```python
# Here we are changing the name of respective species as 1 / 2 / 3 so that model can be applied properly. 
```

```python
y[y=='Iris-setosa']= 1
y[y=='Iris-versicolor']= 2
y[y=='Iris-virginica']= 3
```

```python
#   NOW THE ABOVE SPECIES ARE CHANGED TO : 
#         Iris-setosa  : 1 
#         Iris-versicolor  : 2
#         Iris-virginica  : 3 
     
```

```python
y
```

## Training and testing the data

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20)
```

## Applying Decision Tree  Algorithm

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
```

## Predicting the data from model

```python
y_pred = dt.predict(xtest)
y_pred
```

```python
predicted_data = pd.DataFrame({'Actual result':ytest,'Predicted result':y_pred})
predicted_data
```

## Visualization of model

```python
from sklearn import tree
plt.figure(dpi=300)
tree.plot_tree(dt,fontsize=4)
```

## Confusion matrix and Classification report of model

```python
from sklearn import metrics 
```

```python
confusion_matrix = metrics.confusion_matrix(ytest,y_pred)
confusion_matrix
```

```python
plt.figure(dpi=120)
sns.heatmap(confusion_matrix,annot=True,cmap=plt.cm.Blues)
plt.title('Heat-map of confusion matrix')
plt.xlabel('Predicted values')
plt.ylabel('Actual values')
```

```python
print(metrics.classification_report(ytest,y_pred))
```

## Accuracy of model is 93 %
